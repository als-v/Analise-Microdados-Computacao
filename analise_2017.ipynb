```python
import pandas as pd
```

```python
colunas = ['CO_CATEGAD','CO_GRUPO','CO_REGIAO_CURSO','CO_UF_CURSO','TP_SEXO','TP_PRES','NT_GER','NT_FG','NT_OBJ_FG','NT_DIS_FG','NT_CE','NT_OBJ_CE','NT_DIS_CE','QE_I02','QE_I04','QE_I05','QE_I08','QE_I13','QE_I15','QE_I17','NU_IDADE','CO_CURSO','CO_MODALIDADE','ANO_FIM_EM','ANO_IN_GRAD','CO_TURNO_GRADUACAO','CO_IES','CO_ORGACAD','QE_I10','CO_RS_I9','QE_I22','QE_I23'] #selecionando apenas as colunas necessarias

microdados = pd.read_csv('microdados_Enade_2017_portal_2018.10.09/3.DADOS/MICRODADOS_ENADE_2017 (c√≥pia).csv',sep=';',usecols=colunas,decimal=',') #carregando arquivo, apenas com as colunas desejadas, usando como separador ';'
```

```python
microdados2 = microdados.query('CO_GRUPO == 72 | CO_GRUPO == 79 | CO_GRUPO == 4003 | CO_GRUPO == 4004 | CO_GRUPO == 4005 | CO_GRUPO == 4006 | CO_GRUPO == 6409') #filtrando alunos de cursos da area de TI
```

```python
microdados2 = microdados2.loc[microdados2.TP_PRES.notnull()]
microdados2 = microdados2.astype({'TP_PRES':int}) #convertendo tipo dos dados
```

```python
microdados2 = microdados2.query('TP_PRES == 555') #selecionando apenas alunos presentes no ENADE com nota valida
```

```python
microdados2 = microdados2.loc[microdados2.QE_I15.notnull()] #removendo alunos que deixaram questionario em branco
microdados2 = microdados2.loc[microdados2.NT_GER.notnull()] #removendo alunos com notas em branco
```

```python
#Transforma√ßao
microdados2['CO_CATEGAD'].replace({1:'IES p√∫blica',2:'IES p√∫blica',3:'IES p√∫blica'},inplace=True) #publica 
microdados2['CO_CATEGAD'].replace({4:'IES privada',5:'IES privada',7:'IES privada'},inplace=True) #privada
microdados2 = microdados2.astype({'CO_CATEGAD':str})
```

```python
#agrupando faixas de renda
microdados2['QE_I08'].replace({'B':'A'},inplace=True) #ate 3 salarios
microdados2['QE_I08'].replace({'C':'B','D':'B','E':'B'},inplace=True) #3 a 10 salarios 
microdados2['QE_I08'].replace({'F':'C'},inplace=True) #10 a 30 salarios 
microdados2['QE_I08'].replace({'G':'D'},inplace=True) #mais de 30 salarios 
```

```python
#separando por tipo de ensino medio
microdados2['QE_I17'].replace({'D':'A'},inplace=True) #todo tempo ou maior parte pub
microdados2['QE_I17'].replace({'E':'B'},inplace=True) #todo tempo ou maior parte privada
microdados2['QE_I17'].replace({'F':'C'},inplace=True) #todo tempo ou maior parte exterior
```

```python
estado_dict = {
11 : 'RO',12 : 'AC', 
13 : 'AM',14 : 'RR',
15 : 'PA',16 : 'AP',
17 : 'TO',21 : 'MA',
22 : 'PI',23 : 'CE',
24 : 'RN',25 : 'PB',
26 : 'PE',27 : 'AL',
28 : 'SE',29 : 'BA',
31 : 'MG',32 : 'ES',
33 : 'RJ',35 : 'SP',
41 : 'PR',42 : 'SC',
43 : 'RS',50 : 'MS',
51 : 'MT',52 : 'GO', 
53 : 'DF'
}

microdados2['CO_UF_CURSO'].replace(estado_dict,inplace=True)
microdados2 = microdados2.astype({'CO_UF_CURSO':str})
```

```python
microdados2 = microdados2.astype({'TP_SEXO':str})
microdados2 = microdados2.astype({'ANO_FIM_EM':int})
microdados2 = microdados2.astype({'ANO_IN_GRAD':int})
microdados2 = microdados2.astype({'CO_TURNO_GRADUACAO':int})
microdados2 = microdados2.astype({'TP_PRES':int})
microdados2 = microdados2.astype({'QE_I02':str})
microdados2 = microdados2.astype({'QE_I04':str})
microdados2 = microdados2.astype({'QE_I05':str})
microdados2 = microdados2.astype({'QE_I08':str})
microdados2 = microdados2.astype({'QE_I13':str})
microdados2 = microdados2.astype({'QE_I15':str})
microdados2 = microdados2.astype({'QE_I17':str})
```

```python
microdados2['CO_ORGACAD'].replace({10028:'Universidade',10020:'Centro Universit√°rio',10022:'Faculdade',10026:'I.F.E', 10019:'C.F.E.T'},inplace=True)
```

```python
microdados2['QE_I10'].replace({'A':'N√£o','B':'Sim','C':'Sim','D':'Sim','E':'Sim'},inplace=True)
```

```python
microdados2['QE_I22'].replace({'A':'0','B':'1/2','C':'3-5','D':'6-8','E':'8+'},inplace=True)
```

```python
microdados2['QE_I23'].replace({'A':'0','B':'<3','C':'3-8','D':'8+','E':'8+'},inplace=True)
```

```python
coluna_map = {
    'CO_CATEGAD':'categoria_adm',
    'CO_GRUPO':'grupo',
    'CO_CURSO':'curso',
    'CO_MODALIDADE':'ead_presencial',
    'CO_UF_CURSO':'uf',
    'CO_REGIAO_CURSO':'regiao',
    'NU_IDADE':'idade',
    'TP_SEXO':'sexo',
    'ANO_FIM_EM':'formatura_medio',
    'ANO_IN_GRAD':'inicio_graduacao',
    'CO_TURNO_GRADUACAO':'periodo',
    'TP_PRES':'presenca',
    'NT_GER':'nota_geral',
    'NT_FG':'nota_formacao_geral',
    'NT_OBJ_FG':'nota_obj_formacao_geral',
    'NT_DIS_FG':'nota_disc_formacao_geral',
    'NT_CE':'nota_componente_especifico',
    'NT_OBJ_CE':'nota_obj_componente_especifico',
    'NT_DIS_CE':'nota_disc_componente_especifico',
    'QE_I02':'cor_raca',
    'QE_I04':'escolaridade_pai',
    'QE_I05':'escolaridade_mae',
    'QE_I08':'renda_familiar',
    'QE_I13':'bolsa_academica',
    'QE_I15':'cotas',
    'QE_I17':'tipo_ensino_medio',
    'CO_IES':'codigo_IES',
    'CO_ORGACAD':'organizacao_academica',
    'QE_I10':'trabalha',
    'QE_I22':'livros_ano',
    'QE_I23':'horas_estudo'
}
```

```python
microdados2['grupo'].replace({72: 'ADS', 79:'RC',4003: 'EC', 4004:'BCC',4005:'LCC',4006: 'SI',6409: 'GTI'},inplace=True)
microdados2 = microdados2.astype({'grupo':str})
```

```python
microdados2['cor_raca'].replace({'A':'Branca','B':'Preta','C':'Amarela','D':'Parda','E':'Ind√≠gena','F':'ND'},inplace=True)
```

```python
microdados2['recebeu_bolsa'] = microdados2['bolsa_academica']
microdados2['recebeu_bolsa'].replace({'C':'B','D':'B','E':'B','F':'B'},inplace=True)
microdados2['recebeu_bolsa'].replace({'A':'N√£o','B':'Sim'},inplace=True)
```

```python
microdados2['escolaridade_pai'].replace({'A':'Nenhuma','B':'1¬∞-5¬∞ EF','C':'6¬∞-9¬∞ EF','D':'Ensino M√©dio','E':'Gradua√ß√£o','F':'P√≥s-gradua√ß√£o'},inplace=True)
microdados2['escolaridade_mae'].replace({'A':'Nenhuma','B':'1¬∞-5¬∞ EF','C':'6¬∞-9¬∞ EF','D':'Ensino M√©dio','E':'Gradua√ß√£o','F':'P√≥s-gradua√ß√£o'},inplace=True)
```

```python
# microdados_sul['NU_IDADE'].replace()
bins = [17, 25, 30, 40, 50, 100]
labels = ['17-24', '25-29', '30-39', '40-49', '50+']
microdados2['range_idade'] = pd.cut(microdados2['idade'], bins, labels = labels,include_lowest = True)

# precisa ser algo padronizado, por faixas, n√£o hard code
# ex: lambda <20 = 'A'; 20-30 = 'B' ...
```

```python
microdados2['ead_presencial'].replace({0:'EAD',1:'Presencial'},inplace=True)
microdados2 = microdados2.astype({'ead_presencial':str})
```

```python
microdados2['turno'] = microdados2['periodo'].replace({1:'M',2:'V',3:'I',4:'N'})
microdados2 = microdados2.astype({'turno':str})
```

```python
microdados2['regiao'].replace({1:'N',2:'NE',3:'SE',4:'S',5:'CO'},inplace=True)
microdados2 = microdados2.astype({'regiao':str})
```
